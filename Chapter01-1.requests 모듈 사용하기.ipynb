#### **requests모듈**
  1. http request/response를 위한 모듈
  2. HTTP method를 메서드 명으로 사용하여 request요청  ex)get, post

```python
import requests
```

#### **get 요청하기**
  1. http get 요청하기

```python
url = 'https://news.v.daum.net/v/20191001135734458'
html = requests.get(url)
html.text
```

#### **HTTP response 처리하기**

```python
url = 'https://news.v.daum.net/v/20191001135734458'
html = requests.get(url)

if html.status_code == 200:
    print('크롤링 시작!!')
else:
    print('페이지를 찾을 수 없습니다.')
    print('크롤링 종료')
```
