### **loop(반복문)**
  + 반복적인 작업을 가능하게 해준다
  + 특정 조건을 만족했을 때 반복
  + for : 반복횟수가 정해져 있을 때
  + while : 반복횟수가 정해져 있지 않을 때
  + while은 반드시 반복을 멈추게 하는 장치가 필요! 

#### **while 키워드**
  - 조건이 true가 들어오면 무한루프
  - 조건이 false가 들어오면 반복을 멈춤

```python
a = [5,20,14,27]

i = 0 # 인덱스

while i < len(a):
    print('value: ',a[i],', index', i)
    i += 1
print('ㅋㅋㅋ')
j = 0
for value in a:
    print('value: ',value,', index', j)
    j += 1
```

```python
# 조건문과 함께 사용하기

a = [1,10,9,24,25,26]

i = 0 #인덱스
while i < len(a):
    if a[i] > 20: # 20보다 큰 경우만 출력
        print(a[i])
    i+= 1
```

```python
## false조건: 데이터타입의 기본값들이 오는 경우는 False 반환
# 실제값 % 2 == 0(False) => 짝수

a = [1,10,9,24,25,26]

i = 0
while i < len(a):
    if a[i] %2: # 홀수인 경우에만 출력
        print(a[i])
    else:
        print(a[i]/2)
    i += 1
```

### **무한 루프**
  - while의 경우 true를 조건으로 주면 무한루프

```python
while True:
    print('무한루프다!!')
```

#### **break**
  - loop를 중단할 때 사용
  - loop를 중단하는 경우, while 코드 다음 코드를 실행

```python
a = [1,10,9,24,25,26]

i = 0
while i < len(a): # : 블록 시작
    if a[i] > 20:
        break
    print(a[i])
    i+=1
print('하핳')

num = 10
if num > 5:
    print('111') # 블록문 안에 들어올 내용은 반드시 들여쓰기
    print('222')
print('블록 빠져나가기') # 블록끝 = 들여쓰기를 끝내면
```

#### **continue**
  - break와는 달리 반복문을 종료하지 않고 바로 다음반복으로 넘어가게함

```python
a = 7
while a > 0:
    a -= 1
    if a == 5:
        continue
    print(a)
```

```python
# 1 ~ 100까지 더하는 코드를 작성해보세요
num = 1
_sum = 0

while num <= 100:
    _sum += num
    num += 1
print(_sum)
```
